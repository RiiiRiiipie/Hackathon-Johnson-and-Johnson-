```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

ridge_with_intercept = Ridge(alpha=0.01, fit_intercept=True)
ridge_with_intercept.fit(X_train, y_train)
y_pred_ridge_with_intercept = ridge_with_intercept.predict(X_test)

ridge_without_intercept = Ridge(alpha=0.01, fit_intercept=False)
ridge_without_intercept.fit(X_train, y_train)
y_pred_ridge_without_intercept = ridge_without_intercept.predict(X_test)

lasso_with_intercept = Lasso(alpha=0.01, fit_intercept=True, max_iter=10000)
lasso_with_intercept.fit(X_train, y_train)
y_pred_lasso_with_intercept = lasso_with_intercept.predict(X_test)

lasso_without_intercept = Lasso(alpha=0.01, fit_intercept=False, max_iter=10000)
lasso_without_intercept.fit(X_train, y_train)
y_pred_lasso_without_intercept = lasso_without_intercept.predict(X_test)

ridge_r2_with_intercept = r2_score(y_test, y_pred_ridge_with_intercept)
ridge_r2_without_intercept = r2_score(y_test, y_pred_ridge_without_intercept)

lasso_r2_with_intercept = r2_score(y_test, y_pred_lasso_with_intercept)
lasso_r2_without_intercept = r2_score(y_test, y_pred_lasso_without_intercept)

print("Ridge R^2 With Intercept:", ridge_r2_with_intercept)
print("Ridge R^2 Without Intercept:", ridge_r2_without_intercept)

print("Lasso R^2 With Intercept:", lasso_r2_with_intercept)
print("Lasso R^2 Without Intercept:", lasso_r2_without_intercept)
```
